```python
#Remove punctuation:
import string
def punctuation_removal(text):
    all_list = [char for char in text if char not in string.punctuation]
    clean_str = ''.join(all_list)
    return clean_str
df["content"] = df["content"].apply(punctuation_removal)
df
```

```python
## visualisation
# Most frequent words counter (Code adapted from https://www.kaggle.com/rodolfoluna/fake-news-detector)   
from nltk import tokenize

token_space = tokenize.WhitespaceTokenizer()

def counter(text, column_text, quantity):
    all_words = ' '.join([text for text in text[column_text]])
    token_phrase = token_space.tokenize(all_words)
    frequency = nltk.FreqDist(token_phrase)
    df_frequency = pd.DataFrame({"Word": list(frequency.keys()),
                                   "Frequency": list(frequency.values())})
    df_frequency = df_frequency.nlargest(columns = "Frequency", n = quantity)
    plt.figure(figsize=(12,8))
    ax = sns.barplot(data = df_frequency, x = "Word", y = "Frequency", color = 'blue')
    ax.set(ylabel = "Count")
    plt.xticks(rotation='vertical')
    plt.show()
```

```python
stop = stopwords.words('english')
df["text"] = df["text"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
# Most frequent words in fake news
counter(df[df["Target"] == 0], "text", 20)
```

```python
# trying Ngrams
import re
def basic_clean(text):
  """
  A simple function to clean up the data. All the words that
  are not designated as a stop word is then lemmatized after
  encoding and basic regex parsing are performed.
  """
  wnl = nltk.stem.WordNetLemmatizer()
  stopwords = nltk.corpus.stopwords.words('english')
  words = re.sub(r'[^\w\s]', '', text).split()
  return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

```python
nltk.download('wordnet')
true_word = basic_clean(''.join(str(true['text'].tolist())))
```

```python
true_bigrams_series = (pd.Series(nltk.ngrams(true_word, 2)).value_counts())[:20]
```

```python
true_bigrams_series.sort_values().plot.barh(color='blue', width=.9, figsize=(12, 8))
plt.title('20 Most Frequently Occuring Bigrams')
plt.ylabel('Bigram')
plt.xlabel('# of Occurances')
```

```python
true_trigrams_series = (pd.Series(nltk.ngrams(true_word, 3)).value_counts())[:20]
true_trigrams_series.sort_values().plot.barh(color='red', width=.9, figsize=(12, 8))
plt.title('20 Most Frequently Occuring Trigrams')
plt.ylabel('Trigram')
plt.xlabel('# of Occurances')
```

## Modelling

```python
#split data
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df['content'], df.Target, test_size=0.2, random_state=99)
```

## Logistic Regression

```python
# Vectorizing and applying TF-IDF
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics import accuracy_score

pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', LogisticRegression())])
# Fitting the model
model_LR = pipe.fit(X_train, y_train)
# Accuracy
prediction_LR = model_LR.predict(X_test)
print("accuracy: {}%".format(round(accuracy_score(y_test, prediction_LR)*100,2)))
```

```python

print('---------------------------------------')

print(f'Test score: {model_LR.score(X_test, y_test):.2%}')
print('---------------------------------------')
```

```python
y_test_KPCAp = model_LR.predict(X_test)
y_testp__KPCAlgp = model_LR.predict_proba(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_KPCAp))
```

## Decision Tree Classifier:

```python
from sklearn.tree import DecisionTreeClassifier
# Vectorizing and applying TF-IDF
pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', DecisionTreeClassifier(criterion= 'entropy',
                                           max_depth = 20, 
                                           splitter='best', 
                                           random_state=88))])
# Fitting the model
model_DT = pipe.fit(X_train, y_train)
# Accuracy
prediction_DT = model_DT.predict(X_test)
print("accuracy: {}%".format(round(accuracy_score(y_test, prediction_DT)*100,2)))
```

```python
y_test_KPCAp = model_DT.predict(X_test)
y_testp__KPCAlgp = model_DT.predict_proba(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_KPCAp))
```

## Random Forest Classifier:

```python
from sklearn.ensemble import RandomForestClassifier
pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', RandomForestClassifier(n_estimators=50, criterion="entropy"))])
model_RF = pipe.fit(X_train, y_train)
prediction_RF = model_RF.predict(X_test)
print("accuracy: {}%".format(round(accuracy_score(y_test, prediction_RF)*100,2)))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(list(y_test), prediction_RF)
plt.figure(figsize = (10, 10))
sns.heatmap(cm, annot = True,fmt='g',cmap='viridis')
```

```python
y_test_KPCAp = model_RF.predict(X_test)
y_testp__KPCAlgp = model_RF.predict_proba(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_KPCAp))
```

```python
#GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingClassifier
pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', GradientBoostingClassifier(loss = 'deviance',
                                                   learning_rate = 0.01,
                                                   n_estimators = 10,
                                                   max_depth = 5,
                                                   random_state=55))])

model_GBC = pipe.fit(X_train, y_train)
prediction_GBC = model_GBC.predict(X_test)
print("accuracy: {}%".format(round(accuracy_score(y_test, prediction_GBC)*100,2)))
GBCmodel_accuracy = round(accuracy_score(y_test, prediction_GBC)*100,2)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(list(y_test), prediction_GBC)
plt.figure(figsize = (5, 5))
sns.heatmap(cm, annot = True,fmt='g',cmap='viridis')
```
